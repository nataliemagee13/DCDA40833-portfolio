```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv("politics_data.csv")

# Show the data
print(df)

# Create the chart
plt.figure()
plt.plot(df["Year"], df["Democrat"])
plt.plot(df["Year"], df["Republican"])
plt.plot(df["Year"], df["Independent"])

plt.title("Political Party Support Over Time")
plt.xlabel("Year")
plt.ylabel("Support (%)")
plt.legend(["Democrat", "Republican", "Independent"])

plt.show()
```
